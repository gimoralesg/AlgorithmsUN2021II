```python
import math
# 1. One microsecond by Instructionmath.sqrt(60*10**6)
```

```python
#nlog(n)

def nlogn(s):
  n=2
  maxval = s*(10**6)
  nlogn = 0
  while(nlogn < maxval):
    n = maxval/math.log2(n)
    nlogn += (maxval-nlogn)/2
  return n
```

```python
print("Second = ", nlogn(1))
print("Minute = ", nlogn(60))
print("Hour = ", nlogn(3600))
print("Day = ", nlogn(86400))
print("Year = ", nlogn(31536000))
print("Century = ", nlogn(3154000000))
print("Millenum = ", nlogn(31540000000))
```

```python
#n!
def nfactorial(s):
  n=0
  maxval = s*(10**6)
  nfac = 0
  while(nfac < maxval):
    n += 1
    nfac = math.factorial(n)
  return n-1
```

```python
print("Second = ", nfactorial(1))
print("Minute = ", nfactorial(60))
print("Hour = ", nfactorial(3600))
print("Day = ", nfactorial(86400))
print("Year = ", nfactorial(31536000))
print("Century = ", nfactorial(3154000000))
print("Millenum = ", nfactorial(31540000000))
```

```python
#n^n
def npown(s):
  n=0
  maxval = s*(10**6)
  npown = 0
  while(npown < maxval):
    n += 1
    npown = n**n
  return n-1
```

```python
print("Second = ", npown(1))
print("Minute = ", npown(60))
print("Hour = ", npown(3600))
print("Day = ", npown(86400))
print("Year = ", npown(31536000))
print("Century = ", npown(3154000000))
print("Millenum = ", npown(31540000000))
```
